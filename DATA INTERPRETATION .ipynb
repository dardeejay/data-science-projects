```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime

df = pd.read_csv('student_health.csv')

df
```

```python
df['medical_conditions'].fillna('None', inplace = True)

df = df.filter(regex='^(?!Unnamed:).', axis=1)
df
```

```python
df['allergies'].fillna('None', inplace = True)
df.rename(columns={'height(cm)': 'height', 'weight(kg)':'weight'}, inplace = True)
df
```

```python
gender_count = df.groupby('gender').count()
gender_count
```

```python
gender_count_mean = gender_count.mean(axis=1)
gender_count_mean
```

```python
plt.title('Gender Distribution of Students')
plt.xlabel("Genders")
plt.ylabel("Count")
plt.bar(gender_count.index, gender_count_mean)
```

```python
df
```

```python
df.groupby('category').count().mean(axis=1)
```

```python
plt.title('Distribution of Students BMI Classification')

plt.pie(df.groupby('category').count().mean(axis=1), labels=df.groupby('category').count().index, autopct='%1.1f%%')
```

```python
plt.title('Scatterplot Between the Weight and the Height of Students')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.scatter(df['height'],df['weight'])
```

```python
df
```

```python
 df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
df
```

```python
df['current_year'] = datetime.datetime.now().year

df['age'] = df['current_year'] - df['date_of_birth'].dt.year

age = df.groupby('age').count()
age_mean = age.mean(axis=1)
age_mean

plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Count')
plt.bar(age.index, age_mean)
```

```python
blood_type = df.groupby('blood_type').count()
blood_type_mean = blood_type.mean(axis=1)
plt.title('Blood Type Distrbution of Students')
plt.pie(blood_type_mean, labels = blood_type.index, autopct='%1.1f%%')
```
