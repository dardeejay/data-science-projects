```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('student_info_dup.csv')
```

```python
df.columns
df
```

```python
df.rename(columns={'weight(kg)':'weight','height(cm)':'height'}, inplace=True)
```

```python
df['weight'].fillna(df['weight'].median(), inplace = True)
df['height'].fillna(df['height'].median(), inplace = True)
```

```python

df['weight'].median()
```

```python
plt.bar(df['Gender'].unique(),df.groupby('Gender')['Math'].mean())
plt.xlabel('Genders')
plt.ylabel('Math Average')
plt.title('Average Math Grades per Gender groups in GC CCS')
plt.show()
```
